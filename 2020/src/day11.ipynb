```python
def get_surrounding_occupied_count(df: pd.DataFrame, row_index: int, col_index: int):
    last_row_index = len(df.index) - 1
    top_row_index = 0
    bottom_row_index = 0
    if row_index == 0:
        top_row_index == 0
    else:
        top_row_index = row_index - 1

    if row_index == last_row_index:
        bottom_row_index = last_row_index
    else:
        bottom_row_index = row_index + 1

    last_col_index = len(df.iloc[0]) - 1
    left_col_index = 0
    right_col_index = 0
    if col_index == 0:
        left_col_index == 0
    else:
        left_col_index = col_index - 1
    
    if col_index == last_col_index:
        right_col_index = last_col_index
    else:
        right_col_index = col_index + 1

    area = df.iloc[top_row_index:bottom_row_index + 1, left_col_index:right_col_index + 1]
    
    count_in_area = area.isin(['#']).sum(axis=0).sum()
    if df.iloc[row_index, col_index] == '#':
        return count_in_area - 1
    else:
        return count_in_area

def traverse(row_iterator):
    result = []
    for row_index, row in row_iterator:
            result.append('')
            for col_index, value in row.items():
                #print(col_index, value)
                if value == '.':
                    result[row_index] = result[row_index] + '.'
                elif value == 'L':
                    occupied_count = get_surrounding_occupied_count(input, row_index, col_index)
                    if occupied_count == 0:
                        result[row_index] = result[row_index] + '#'
                    else:
                        result[row_index] = result[row_index] + '#'
                else:
                    occupied_count = get_surrounding_occupied_count(input, row_index, col_index)
                    if occupied_count > 3:
                        result[row_index] = result[row_index] + 'L'
                    else: 
                        result[row_index] = result[row_index] + '#'
    return result

def program():
    import pandas as pd # Import pandas
    input = pd.read_fwf("../inputs/day11.csv", header=None)
    input = input[0].apply(lambda r: pd.Series(list(r)))

    file = open("../inputs/day11.csv")
    previous = file.read() # Read a text file
    previous = previous.split("\n")

    row_iterator = input.iterrows()

    while True:
        # Loop through all values in the grid and compute new value
        result = traverse(row_iterator)
        #print(result)
        
        # Compare the new grid to the old grid
        if result == previous:
            print('Found equal lists')
            print(result, previous)
            return input.isin(['#']).sum(axis=0).sum()
        else:
            print('Differences found')
            new_input = pd.DataFrame(result)
            new_input = new_input[0].apply(lambda r: pd.Series(list(r)))
        row_iterator = new_input.iterrows()
        previous = result

program()
```
